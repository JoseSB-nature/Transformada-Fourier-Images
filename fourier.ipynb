```python
from skimage import io
from skimage.color import rgb2gray
import numpy as np
import matplotlib.pyplot as plt


p=[[]]
```

```python
def furieriza(path_imag):
  plt.axis('off')
  imag = io.imread(path_imag)
  gray_imag = rgb2gray(imag)
  io.imshow(imag)
  plt.figure(num=None, figsize=(8, 6), dpi=80)
  # plt.axis('off')
  # plt.imshow(gray_imag, cmap='gray');
  gray_imag_fourier = np.fft.fftshift(np.fft.fft2(gray_imag))
  plt.figure(num=None, figsize=(8, 6), dpi=100)
  plt.axis('off')

  plt.imshow(np.log(abs(gray_imag_fourier)) ,interpolation='spline16', cmap='magma')
  #print(np.log(abs(gray_imag_fourier)))
```
